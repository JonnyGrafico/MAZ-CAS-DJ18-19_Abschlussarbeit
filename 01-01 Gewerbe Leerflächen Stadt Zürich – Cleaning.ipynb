## Datensatz säubern
Die Angaben zu der Art der Leerflächen, sowie die Jahreszahlen liegen in unterschiedlich verteilten "rows" vor.

Diese müssen nun jeweils in eine einzelne "row" pro "column" zusammengeführt werden damit eine saubere Selektion der gewünschten Daten gelingt.

Zudem sind diverse Textelemente zu eliminieren. Damit die Daten als reine Zahlen erscheinen muss der Datensatz in int64 konvertiert werden, ansonsten werden sie mit Nachkommastellen dargestellt und die Auswertung funktioniert nicht.

### Rows verbinden für neuen Header

```python
# looking at original data, setting header to row 7
df1 = pd.read_excel('Daten/Gewerbe/Leerflächen_Gewerbe_Art_Quartier.xlsx', header=7).replace('–', np.nan)
df1.head()
```

```python
# making a list of df1
list1=list(df1)
```

```python
# looking at original data, setting header to row 9
df2 = pd.read_excel('Daten/Gewerbe/Leerflächen_Gewerbe_Art_Quartier.xlsx', header=9).replace('–', np.nan)
df2.head()
```

```python
# resetting index to same colmuns lenght as df1
df2 = df2.reset_index()
df2.head()
```

```python
# making a second list of df2
list2=list(df2)
```

```python
# looking at original data, setting header to row 9
df3 = pd.read_excel('Daten/Gewerbe/Leerflächen_Gewerbe_Art_Quartier.xlsx', header=10).replace('–', np.nan)
df3.head()
```

```python
# resetting index to same colmuns lenght as df1 and df2
df3 = df3.reset_index().head()
```

```python
# making a third list of df3
list3=list(df3)
```

#### list1 + list2 + list3 = list4 is new header

```python
# creating an new list4 with data and header info of list1, list2 and list3 
list4 = []
for elem1, elem2, elem3 in zip(list1, list2, list3):
    elem = elem1, elem2, elem3
    list4.append(elem)
```

```python
# looking at original data, setting header to row 7
df = pd.read_excel('Daten/Gewerbe/Leerflächen_Gewerbe_Art_Quartier.xlsx', header=7).replace('–', np.nan)
```
