### Rows verbinden für neuen Header

Daten säubern – Rows für Header verbinden

#### 1. Liste

```python
# looking at original data, setting header to row 7
df1 = pd.read_excel('Daten/Wohnungen/00_Leerwohnungen-Stadt-Zuerich_Original.xlsx', header=7
                  ).replace('–', np.nan).fillna(0)
df1.head()
```

```python
# making a list of df1
list1=list(df1)
```

#### 2. Liste

```python
# looking at original data, setting header to row 8
df2 = pd.read_excel('Daten/Wohnungen/00_Leerwohnungen-Stadt-Zuerich_Original.xlsx', header=8
                  ).replace('–', np.nan).fillna(0)
df2.head()
```

```python
# making a list of df2
list2=list(df2)
```

#### 1. Liste + 2. Liste = 3. Liste mit neuem Header

```python
# creating an new list3 with data and header info of list1, list2 
list3 = []
for elem1, elem2 in zip(list1, list2):
    elem = elem1, elem2
    list3.append(elem)
```

```python
# looking at original data, setting header to row 7
df = pd.read_excel('Daten/Wohnungen/00_Leerwohnungen-Stadt-Zuerich_Original.xlsx'
                   ,index_col=None, header=7).replace('–', np.nan).fillna(0)
```

```python
# replacing columns with new created list3
df.columns = list3
df.head()
```
