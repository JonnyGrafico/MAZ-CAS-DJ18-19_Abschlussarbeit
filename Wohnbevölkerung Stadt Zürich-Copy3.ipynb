# Ständige Wohnbevölkerung Stadt Zürich – 1993 bis 2017

```python
df = pd.read_excel('Daten/Bevölkerung/Ständige-Wohnbevölkerung.xlsx', header=7).fillna(0)
```

```python
df.head()
```

```python
df.columns
```

```python
trace1 = go.Scatter(x=df.Jahr, y=df['Ständige Wohnbevölkerung'], 
                    name="Ständige Wohnbevölkerung",
                    mode='lines+markers',
# mode can be 'markers', 'lines+markers', 'lines'
                    
                    fill='tozeroy',
                    
                        line = dict(
                            width = 4,
                            color = '#5120B3',
                        ),
                    
                        marker = dict(
                            symbol='circle-dot',
                            size = 10,
                            color = '#84B5F4',
                            line = dict(
                                width = 2
                            )
                        )
                    )  

data = [trace1]

layout = go.Layout(
    title='Ständige Wohnbevölkerung – 1993 bis 2018',

    xaxis=dict(
    ),
    yaxis=dict(
        rangemode='tozero',
        nticks=8,
        exponentformat='none'
    )
)
fig = go.Figure(data=data, layout=layout)
pyo.iplot(fig, filename='hover-chart-basic')




# data = [trace0, trace1, trace2,trace3]  # assign traces to data
```

# Ständige Wohnbevölkerung – 2010 bis 2017

```python
df2 = df[['Jahr', 'Ständige Wohnbevölkerung']][17:]
```

```python
df2
```

```python
trace1 = go.Scatter(x=df2.Jahr, y=df2['Ständige Wohnbevölkerung'], 
                    name="Ständige Wohnbevölkerung",
                    mode='lines+markers',
                    # mode can be 'markers', 'lines+markers', 'lines'
                    fill='tozeroy',
                    
                        line = dict(
                            width = 4,
                            color = '#5120B3'
                        ),
                        marker = dict(
                            symbol='circle-dot',
                            size = 10,
                            color = '#84B5F4',
                            line = dict(
                                width = 2)
                        )
                    ) 

data = [trace1]



layout = go.Layout(
    title='Ständige Wohnbevölkerung – 2010 bis 2018',
    font=dict(
            family='arial',
            size=14,
            color='#7f7f7f'),

    xaxis=dict(
        title='Jahr',
        titlefont=dict(
            family='Courier New, monospace',
            size=14,
            color='#7f7f7f')
    ),
    
    yaxis=dict(
#         title='Anzahl Personen',
        title=dict(
            text='Anzahl Personen',
            ),
        
        titlefont=dict(
            family='Courier New, monospace',
            size=14,
            color='#7f7f7f'),
        rangemode='tozero',
        nticks=8,
        exponentformat='none',
        tickwidth=1,
    )
)
fig = go.Figure(data=data, layout=layout)
pyo.iplot(fig, filename='hover-chart-basic')
```
